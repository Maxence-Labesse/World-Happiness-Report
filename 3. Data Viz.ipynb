```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import requests

from utils.Dataviz_utils import *
from utils.Display_utils import print_title1
```

We import and prepare the data

```python
df = pd.read_csv('data/data_clean_norm.csv')
print("data dimension:",df.shape)

# Happiness quartiles
df["happiness_level"]=pd.qcut(df["happiness_score"], 4, labels=["low", "low-mid", "top-mid", "top"])
l_level = ["top", "top-mid", "low-mid", "low"]

# years to string
df["year"] = df["year"].apply(lambda x: str(x))

# renaming some countries to match the map template
df['country'].replace({'Serbia': 'Republic of Serbia', 'Tanzania': 'United Republic of Tanzania',
                       'Congo (Brazzaville)': 'Republic of the Congo', 'Congo (Kinshasa)': 'Democratic Republic of the Congo',
                       'United States': 'United States of America'}, inplace=True)


# dataset containing data for 2019
df_2019 = df.loc[df["year"]=="2019"]

# dataset_containing data for 2015
df_2015 = df.loc[df["year"]=="2015"].copy()
df_2015.rename(columns=dict(zip(df.columns.tolist(),[col+"_2015" if col!="country" else col for col in df.columns.tolist()])), inplace=True)

# list containing the 6 factors names
l_factors = ['life_expectancy', 'gdp_per_capita', 'social_support', 
             'freedom','generosity', 'corruption_perception'] 
```

We prepare template data to build maps

```python
# import world country boundaries
url = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/world_countries.json'
response = requests.get(url)
data = response.json()
  
# template to build map
world_geo = r'world_countries.json'
```

## Which factors explain the differences in happiness ?

```python
corr_matrix=df_2019[["happiness_score"]+l_factors].corr()

sns.heatmap(corr_matrix, annot = True)
```

The correlation matrix gives us an idea of the strength of the relationships between `happiness_score` and the different factors: With correlation >0.7, a relationship is considered as strong, while one with <0.3 is considered as weak.

`Happiness_score` seems to be strongly correlated to `gdp per capita`, `life expectancy`, `social supports` and, to a lesser extent, `freedom`. Let's try to explore the nature of these relationships

### Life expectancy, gdp per capita, and social support

```python
g=sns.pairplot(df_2019, y_vars="happiness_score", x_vars=['life_expectancy', 'social_support','gdp_per_capita'], hue="happiness_level",
            palette=['#F6EFF7', '#A6BDDB', '#3690C0', '#016450'], diag_kind = None)
g.fig.suptitle("Relationship between happiness and the 3 most correlated factors", y=1.08)
g._legend.set_title("Happiness")
```

Looking at the distribution of the countries in respect to happiness and the 3 most correlated factors, we can observe that these relationships are very linear: **the higher the factors, the higher the happiness**

### Freedom

```python
numeric_th_graph(df_2019, "happiness_level", "freedom", 60, "under")
```

Although it has an average correlation with `happiness` (0.56), a high `freedom` level seems to be a requirement to feel happy in a country. We can observe this on the graphs above, as **about 3 countries with a `freedom` perception <60 in 4 have a low/low-mid level of happiness**

### Corruption perception

```python
numeric_th_graph(df_2019, "happiness_level", "corruption_perception", 60, "over")
```

Despite a weak correlation with `happiness`, we can observe a very meaningful behavior here.
On the left plot, we can see that most of the countries for with high `coruption perception` belong to the TOP25% happiest countries. 
Precisely, the chart on the right evidences that **almost 85% of the countries with `coruption perception` >60 belong to the TOP happiest countries**.

### Generosity

```python
g=sns.pairplot(df_2019, y_vars="happiness_score", x_vars=['generosity'], hue="happiness_level",
            palette=['#F6EFF7', '#A6BDDB', '#3690C0', '#016450'], diag_kind = None)
g.fig.suptitle("Relationship between happiness and generosity", y=1.08)
g._legend.set_title("Happiness")
g.fig.set_size_inches(10,4)
```

Like we saw in correlation matrix, `generosity` have a weak (0.1) relationships with `happiness`. There is no particular pattern to highlight here

# 2. Trends in happiness from 2015 to 2019

<a name="In"></a>
## In which countries/regions happiness has increased during this period?

```python
df_evol = pd.merge(df, df_2015, on="country", how="inner")

for col in ['happiness_score']+l_factors:
    df_evol['e_'+ col] = df_evol[col] - df_evol[col+"_2015"]
    
l_evol_factors = ['e_'+col for col in l_factors]
```

```python
country_map(df_evol, "country", "e_happiness_score", [45, 0], 1.8, world_geo, "Happiness evolution from 2015 to 2019",
            threshold_scale=[-45, -10, 0, 5, 35])
```

We can make several observations with the map:
* **Latin America** Happiness has significantly decreased between 2015 and 2019 (-45 to -10)
* **Estearn Asia**, **Western Europe** happiness has increased (+0 to +35)
* **Western Africa** happiness has sharply increased (+5 to +35)

----------
<a name="Ex"></a>
## What does explain this growth?

```python
corr_matrix=df_evol[["e_happiness_score"]+l_evol_factors].corr()

sns.heatmap(corr_matrix, annot = True)
```

The correlation matrix indicates that evolution of happiness is not correlated with the evolution of the factors from 2015 to 2019. Let's see if cross distribution can help us to find more information.

```python
sns.pairplot(df_evol, y_vars="e_happiness_score", x_vars=l_evol_factors,  diag_kind = None)
```

There is no specific relationship to highlight between happiness evolution and factors evolution.

According to the data we have, we assume that we can't explain the evolution of happiness between 2015 and 2019.
